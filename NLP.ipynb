```python
# NATURAL LANGUAGE PROCESSING - NLP
import pandas as pd

import nltk
nltk.download()
```

```python
# ************************************************PREPROCESSING:

# punctuation marks removal:
import string
def punct_rem(phrase):
    clear_p=''.join(char for char in phrase if char not in string.punctuation)
    return clear_p

punct_rem('?abc:=defgh!')
```

```python
# tokenization:
import re
def tokenize(phrase):
    new_t=re.split('\W+', phrase)
    return new_t

pd.Series(tokenize('Ala ma kota.')).apply(lambda x: x.lower())
```

```python
# stopwords removal:

stopword=nltk.corpus.stopwords.words('english')
stopword[:5]
```

```python
def stopword_rem(phrase):
    clear_p=[word for word in phrase if word not in stopword]
    return clear_p
```

```python
# stemming - removal of the suffices like 'ing', 'ly' etc. (the corpus remains untouched)
ps=nltk.PorterStemmer()

def stemming(phrase):
    stemmed=[ps.stem(word) for word in phrase]
    return stemmed
```

Stemming is typically faster as it simply chops off the end of the word, without understanding the context of the word. Lemmatizing is slower and more accurate as it takes an informed analysis with the context of the word in mind.

```python
# lemmatizing:
nltk.download('wordnet')
wn= nltk.WordNetLemmatizer()

def lemm(phrase):
    clear_p=[wn.lemmatize(word) for word in phrase]
    return clear_p
```

```python
# EXAMPLE - SPAM DATASET:

df=pd.read_csv('https://raw.githubusercontent.com/justkacz/csvfiles/main/SMSSpamCollection.tsv',
               sep='\t', names=['label', 'body_text'])
df.head()
```

```python
# punctuation removal
df['punct_rem']=df['body_text'].apply(lambda x: punct_rem(x))
df.head()
```

```python
# tokenization:
df['tokenized']=df['punct_rem'].apply(lambda x: tokenize(x.lower()))
df.head()
```

```python
df['no_stopword']=df.tokenized.apply(lambda x: stopword_rem(x))
df.head()
```

```python
# stemming:
df['stemmed']=df['no_stopword'].apply(lambda x: stemming(x))
df.head()
```

```python
df['lemmatized']=df.no_stopword.apply(lambda x : lemm(x))
df.head()
```
